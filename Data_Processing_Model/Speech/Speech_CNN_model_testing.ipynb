```python
data_dir = "/content/drive/MyDrive/DS340 Final Project/Speech/speech_spectrograms"
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Image dimensions
IMG_HEIGHT, IMG_WIDTH = 128, 128
BATCH_SIZE = 32

# Create data generators
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training'
)

val_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'
)
```

```python
from tensorflow.keras import layers
from tensorflow import keras
input_shape = (128, 128, 3)
```

```python
base_model = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Conv2D(32, kernel_size=(3, 3), activation='relu'),
        layers.Conv2D(64, kernel_size=(3, 3), activation='relu'),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dense(8, activation='softmax')
    ]
)
```

```python
base_model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
base_model.fit(train_generator, epochs=15, validation_data=val_generator)
```

```python
model1 = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Conv2D(32, kernel_size=(3, 3), activation='relu'),
        layers.Conv2D(64, kernel_size=(3, 3), activation='relu'),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Flatten(),
        layers.Dropout(0.5), # add dropout
        layers.Dense(128, activation='relu'),
        layers.Dense(8, activation='softmax'),
    ]
)
```

```python
model1.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
model1.fit(train_generator, epochs=15, validation_data=val_generator)
```

```python
model2 = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Conv2D(32, kernel_size=(3, 3), activation='relu'),
        layers.Conv2D(64, kernel_size=(3, 3), activation='relu'),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Flatten(),
        # remove droput
        layers.Dense(128, activation='relu'),
        layers.Dense(8, activation='softmax'),
    ]
)
```

```python
model2.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

import keras.callbacks
callbacks = [
    keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)  # Quit after 3 rounds of no validation loss improvement
]

model2.fit(train_generator, epochs=15, validation_data=val_generator, callbacks=callbacks)# add early stopping
```

```python
model5 = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Conv2D(32, kernel_size=(3, 3), activation='relu'),
        layers.MaxPooling2D(pool_size=(2, 2)), # add max pooling
        layers.Conv2D(64, kernel_size=(3, 3), activation='relu'),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(256, (3, 3), activation='relu'), # add another layer
        layers.MaxPooling2D(pool_size=(2, 2)), # add max pooling
        layers.Flatten(),
        layers.Dropout(0.5),
        layers.Dense(128, activation='relu'),
        layers.Dense(8, activation='softmax'),
    ]
)
```

```python
model5.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

import keras.callbacks
callbacks = [
    keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)  # Quit after 3 rounds of no validation loss improvement
]

model5.fit(train_generator, epochs=15, validation_data=val_generator, callbacks=callbacks)
```

```python
from keras import regularizers
from keras import layers, regularizers, Sequential

model6 = Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Conv2D(32, kernel_size=(3, 3), activation='relu', kernel_regularizer=regularizers.l2(1e-4)), # add l2 regularization
        # remove max pooling
        layers.Conv2D(64, kernel_size=(3, 3), activation='relu', kernel_regularizer=regularizers.l2(1e-4)), #add l2 regularization
        layers.MaxPooling2D(pool_size=(2, 2)),

        layers.Conv2D(128, (3, 3), activation='relu', kernel_regularizer=regularizers.l2(1e-4)), #add l2 regularization
        # remove max pooling
        layers.Conv2D(256, (3, 3), activation='relu', kernel_regularizer=regularizers.l2(1e-4)), #add l2 regularization
        layers.MaxPooling2D(pool_size=(2, 2)),

        layers.Flatten(),
        layers.Dropout(0.6), # change dropout from 0.5 to 0.6
        layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(1e-4)), #add l2 regularization
        layers.Dense(8, activation='softmax'),
    ]
)
```

```python
model6.compile(loss="categorical_crossentropy", optimizer=keras.optimizers.Adam(learning_rate=1e-4), metrics=["accuracy"]) # modify learning rate

import keras.callbacks
callbacks = [
    keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)  # Quit after 3 rounds of no validation loss improvement
]

model6.fit(train_generator, epochs=15, validation_data=val_generator, callbacks=callbacks)
```

```python
from keras import regularizers
from keras import layers, regularizers, Sequential

model7 = Sequential([
    keras.Input(shape=input_shape),

    layers.Conv2D(32, kernel_size=(3, 3), activation='relu', kernel_regularizer=regularizers.l2(1e-4)),
    layers.Conv2D(64, kernel_size=(3, 3), activation='relu', kernel_regularizer=regularizers.l2(1e-4)),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Dropout(0.3), # add another dropout

    layers.Conv2D(128, (3, 3), activation='relu', kernel_regularizer=regularizers.l2(1e-4)),
    # remove max pooling
    layers.Conv2D(128, (3, 3), activation='relu', kernel_regularizer=regularizers.l2(1e-4)), # change 256 to 128
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Dropout(0.4), # add another dropout

    layers.Flatten(),
    layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(1e-4)),
    layers.Dropout(0.5), # change dropout from 0.6 to 0.5
    layers.Dense(8, activation='softmax'),
])
```

```python
model7.compile(loss="categorical_crossentropy", optimizer=keras.optimizers.Adam(learning_rate=1e-4), metrics=["accuracy"])

import keras.callbacks
callbacks = [
    keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)  # Quit after 3 rounds of no validation loss improvement
]

model7.fit(train_generator, epochs=15, validation_data=val_generator, callbacks=callbacks)
```
